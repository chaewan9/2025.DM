```python
# 전염 시뮬레이션 (BFS 응용)
from collections import deque, defaultdict

def simulate_infection(graph, start_person, max_depth=10):
  visited = set()
  queue = deque([(start_person, 0)])

  print(f"{start_person}의 전염 경로 (최대 10단계) : ")

  # 단계별로 노드 저장
  levels = defaultdict(list)

  while queue:
    current, depth = queue.popleft()
    if current not in visited:
      visited.add(current)
      if depth > 0:
        levels[depth].append(current)
      if depth < max_depth:
        for neighbor in graph[current]:
          if neighbor not in visited:
            queue.append((neighbor, depth + 1))

  for depth in sorted(levels.keys()):
    print(f"{depth}단계 : {', '.join(levels[depth])}")
simulate_infection(graph, 'Son')
```
